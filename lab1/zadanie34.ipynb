```python
import numpy as np
import scipy
import matplotlib
import matplotlib.pyplot as plt

x_1 = np.linspace(-1, 1)
x_2 = np.linspace(-6, 6)
x_3 = np.linspace(0, 5)
y = x_1 ** 2 + 5 
plt.plot(x_1, y, label ='x ** 2 + 5') 
plt.xlabel("argumenty x") 
plt.ylabel("argumenty y")
plt.title("Funkcja 1")
plt.grid() 
plt.legend()
plt.show()
```

```python
plt.plot(x_2, y, label ='x ** 2 + 5') 
plt.xlabel("argumenty x") 
plt.ylabel("argumenty y")
plt.title("Funkcja 2")
plt.grid() 
plt.legend()
plt.show()
```

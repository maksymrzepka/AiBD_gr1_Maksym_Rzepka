```python
import pandas as pd
from matplotlib import pyplot as plt

data = r'C:\Users\Acer\OneDrive\Pulpit\STUDIA1\sem5\AiBD\lab5\Analysis Data/'

# oryginalna tabela z danymi
tb_original_ = pd.read_csv(f"{data}tb_original.csv", sep=';')
print(tb_original_)
```

***Sumuje mezczyzn i kobiety z kazdego przedzialu wiekowego, aby nastepnie stworzyc wykres przedstawiajacy zachorowania wsrod kobiet i mezczyzn*** 

```python
tb = tb_original[['year','iso2']]
sum_male = tb_original[['new_sp_m014', 'new_sp_m1524', 'new_sp_m2534', 'new_sp_m3544', 'new_sp_m4554',
     'new_sp_m5564', 'new_sp_m65', 'new_sp_mu']]
sum_female = tb_original[['new_sp_f014', 'new_sp_f1524', 'new_sp_f2534', 'new_sp_f3544', 'new_sp_f4554',
     'new_sp_f5564', 'new_sp_f65', 'new_sp_fu']]

tb.loc[:, 'male_sickness'] = sum_male.sum(numeric_only=True, axis=1)
tb.loc[:, 'female_sickness'] = sum_female.sum(numeric_only=True, axis=1)

print(tb)

x_1 = tb['year']
y_1 = tb['male_sickness']

x_2 = tb['year']
y_2 = tb['female_sickness']

plt.bar(x_1, y_1, label = 'mezczyzni')
plt.bar(x_2, y_2, label = 'kobiety')
plt.title("Zachorowania wsrod kobiet i mezczyzn")
plt.xlabel("Rok")
plt.xlim([1994, 2009])
plt.ylabel("Liczba chorych kobit i mezczyzn")
plt.legend()
plt.grid()
plt.show()
```
